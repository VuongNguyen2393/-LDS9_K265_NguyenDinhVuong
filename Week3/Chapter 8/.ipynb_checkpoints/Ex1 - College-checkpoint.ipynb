```python
from pyspark import SparkContext
from pyspark.sql import SparkSession
```

```python
sc = SparkContext(master='local',appName='Ex1')
spark = SparkSession(sc)
```

### Read data

```python
data = spark.read.csv('./College.csv',inferSchema=True,header=True)
```

```python
data.printSchema()
```

```python
data.select('School','Private','Apps').show(3)
```

### Pre-processing

###### Check dữ liệu nan, null

```python
from pyspark.sql.functions import col, count, isnan, isnull, when
```

```python
#Check nan
data.select([count(when(isnan(col(x)),col(x))).alias(x) for x in data.columns]).toPandas()
```

```python
#Check null
data.select([count(when(isnull(col(x)),col(x))).alias(x) for x in data.columns]).toPandas()
```

**Nhận xét:** Dữ liệu không có dự liệu Nan, Null

```python
#Check duplicate
```

```python
duplicate_row = data.count() - data.distinct().count()
print('Số dữ liệu bị trùng:', duplicate_row)
```

**Nhận xét:** Dữ liệu không bị trùng

```python
data.toPandas().head(10)
```

##### Scale Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numeric_col = data.drop('School','Private').columns
```

```python
plt.figure(figsize=(20,16))
for i,col in enumerate(numeric_col):
    plt.subplot(5,4,i+1)
    sns.distplot(data.select(col).toPandas())
```

```python
plt.figure(figsize=(20,16))
for i,col in enumerate(numeric_col):
    plt.subplot(5,4,i+1)
    sns.boxplot(data.select(col).toPandas())
```

**Nhận xét:** 
Tập dữ liệu numeric chứa bao gồm thông số lệch trái và phải, chứa outlier nên ta chọn phương pháp Scale là Robust

```python
import numpy as np
```

```python
for col in numeric_col:
    Q1 = np.percentile(data.select(col).toPandas(),75)
    Q3 = np.percentile(data.select(col).toPandas(),25)
    data = data.withColumn(col + '_scale',(data[col] - Q1)/(Q3-Q1))
```

##### Label Encoder đối với cột Private

```python
data = data.withColumn('Private_idx',when(data['Private']=='Yes',1).otherwise(0))
```

### Train - Test data preparation

```python
from pyspark.ml.linalg import Vector
from pyspark.ml.feature import VectorAssembler
```

```python
data.columns
```

```python
assembler = VectorAssembler(inputCols=('Apps_scale',
 'Accept_scale',
 'Enroll_scale',
 'Top10perc_scale',
 'Top25perc_scale',
 'F_Undergrad_scale',
 'P_Undergrad_scale',
 'Outstate_scale',
 'Room_Board_scale',
 'Books_scale',
 'Personal_scale',
 'PhD_scale',
 'Terminal_scale',
 'S_F_Ratio_scale',
 'perc_alumni_scale',
 'Expend_scale',
 'Grad_Rate_scale'),outputCol='features')
```

```python
data = assembler.transform(data)
```

```python
final_data = data.select('features','Private_idx')
```

```python
train_data, test_data = final_data.randomSplit([0.7,0.3])
```

### Classification

```python
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.classification import GBTClassifier
```

```python
tree_clf = DecisionTreeClassifier(featuresCol='features',labelCol='Private_idx')
forest_clf = RandomForestClassifier(featuresCol='features',labelCol='Private_idx')
GBT_clf = GBTClassifier(featuresCol='features',labelCol='Private_idx')
```

```python
#train model
tree_model = tree_clf.fit(train_data)
forest_model = forest_clf.fit(train_data)
GBT_model = GBT_clf.fit(train_data)
```

```python
#prediction:
tree_predict = tree_model.transform(test_data)
forest_predict = forest_model.transform(test_data)
GBT_predict = GBT_model.transform(test_data)
```

### Evaluattion

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
```

```python
acc_evaluate = MulticlassClassificationEvaluator(labelCol='Private_idx',metricName='accuracy')
```

```python
tree_acc = acc_evaluate.evaluate(tree_predict)
forest_acc = acc_evaluate.evaluate(forest_predict)
GBT_acc = acc_evaluate.evaluate(GBT_predict)
```

```python
print('Accuracy of decision tree model: ', tree_acc)
print('Accuracy of Random Forest model: ',forest_acc)
print('Accuracy of Gradient Boost Tree: ',GBT_acc)
```

**Nhận xét :**
Độ chính xác của 3 thuật toán là khá tốt, đặc biệt với các parameter mặc định thì Random Forest cho kết quả vượt trội với việc sử dụng 20 cây
